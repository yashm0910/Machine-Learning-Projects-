```python
#Now we will move to part of feature engineering

X = df.drop('Class', axis=1)
y = df['Class'].values

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_x = scaler.fit_transform(X)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(scaled_x, y, test_size=0.2, random_state=40)

from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100, max_depth=7, random_state=42)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)

```

```python
# Now we will print three important metrics to analyze the model
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
print('''
Precision (Class 1): 0.94 - 94% of predicted frauds were correct
Recall (Class 1): 0.82 - It only caught 82% of actual frauds
=> So it missed 18% frauds!
''')
```

```python
#This project being entry level we will use basic  Decision tree Algorithm to apply linear regression in different model
from xgboost import XGBClassifier

xgb=XGBClassifier(n_estimators=5,max_depth=5,learning_rate=0.1, eval_metric='logloss' ,random_state=42)
xgb.fit(X_train,y_train)
new_y_pred=xgb.predict(X_test)
```

```python
print("Accuracy:", accuracy_score(y_test, new_y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, new_y_pred))
print("Classification Report:\n", classification_report(y_test, new_y_pred))
print('''
Precision (Class 1): 0.92 - Higher than Logistic and Decision Tree, lower than Random forest
Recall (Class 1): 0.78 - lesser than random forest
=> It caught 78% of frauds (only missed 22%)''')
```

```python
model_results = {
    'Model': ['Logistic Regression', 'Decision Tree', 'Random Forest', 'XGBoost'],
    'True Negative': [56857, 56842, 56860, 56858],
    'False Positive': [8, 23, 5, 7],
    'False Negative': [34, 17, 17, 21],
    'True Positive': [63, 80, 80, 76],
    'Precision (Class 1)': [0.89, 0.78, 0.94, 0.92],
    'Recall (Class 1)': [0.65, 0.82, 0.82, 0.78],
    'F1-Score (Class 1)': [0.75, 0.80, 0.88, 0.84],
    'Accuracy': [0.99926, 0.99930, 0.99961, 0.99988]
}
results_df = pd.DataFrame(model_results)
```

```python
results_df['Specificity'] = results_df['True Negative'] / (results_df['True Negative'] + results_df['False Positive'])
results_df['False Positive Rate'] = 1 - results_df['Specificity']
results_df['Fraud Detection Rate'] = results_df['True Positive'] / (results_df['True Positive'] + results_df['False Negative'])


results_df = results_df[['Model', 'True Negative', 'False Positive', 'False Negative', 'True Positive', 
                         'Precision (Class 1)', 'Recall (Class 1)', 'F1-Score (Class 1)', 
                         'Specificity', 'False Positive Rate', 'Fraud Detection Rate', 'Accuracy']]

percentage_cols = ['Precision (Class 1)', 'Recall (Class 1)', 'F1-Score (Class 1)', 
                   'Specificity', 'False Positive Rate', 'Fraud Detection Rate', 'Accuracy']

results_df[percentage_cols] = results_df[percentage_cols].applymap(lambda x: f"{x:.2%}")

results_df
```
